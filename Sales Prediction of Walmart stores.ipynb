```python
data1 =pd.merge(train,features, on=['Store','Date'], how='inner')
data1
```

```python
from datetime import datetime as dt
```

```python
data1['Date_Time_Obj']=[dt.strptime(x,'%Y-%m-%d') for x in list(data1['Date'])]
data1.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(data1[(data1.Store==1)].Date_Time_Obj, data1[(data1.Store==1)].Weekly_Sales,'go')
plt.show()
```

```python
weeklysales=data1.groupby(['Store','Date'])['Weekly_Sales'].apply(lambda x:np.sum(x))
weeklysales[0:10]
```

```python
weeklysale_dept=data1.groupby(['Store','Dept'])['Weekly_Sales'].apply(lambda x:np.sum(x))
weeklysale_dept[0:5]
```

```python
weeklysales_details=weeklysales.reset_index()
weeklysales_details[0:5]
```

```python
walmart_store=pd.merge(weeklysales_details, features, on=['Store', 'Date'], how='inner')
walmart_store.head()
```

```python
walmart_store['Date_Time_Obj']=[dt.strptime(x,'%Y-%m-%d') for x in list(walmart_store['Date'])]
walmart_store.head()
```

```python
walmart_store_new = walmart_store.iloc[:, [0,1,2,3,4,10,11,12,13]]
walmart_store_new = walmart_store_new.reset_index(drop=True)
walmart_store_new
```

```python
Week_Num = [(x - walmart_store_new['Date_Time_Obj'][0]) for x in list(walmart_store_new['Date_Time_Obj'])]
```

```python
walmart_store_new['Week'] = [np.timedelta64(x, 'D').astype(int)/7 for x in Week_Num]
```

```python
walmart_store_new.head()
```

```python
walmart_store_new['IsHolidayInt'] = [int(x) for x in list(walmart_store_new.IsHoliday)]
```

```python
walmart_store_new.head()
```

```python
walmart_store_new.Store.unique()
```

```python
import sklearn
from sklearn.cross_validation import train_test_split
plt.plot(walmart_store_new[(walmart_store_new.Store==1)].Week, walmart_store_new[(walmart_store_new.Store==1)].Weekly_Sales, 'ro')
plt.show()
```

```python
train_Walmart, test_Walmart = train_test_split(walmart_store_new, test_size=0.3,random_state=42)
```

```python
XTrain = train_Walmart[['Temperature', 'Fuel_Price', 'CPI', 'Unemployment', 'Week', 'IsHolidayInt']]
YTrain = train_Walmart['Weekly_Sales']
```

```python
XTest = test_Walmart[['Temperature', 'Fuel_Price', 'CPI', 'Unemployment', 'Week', 'IsHolidayInt']]
YTest = test_Walmart['Weekly_Sales']
```

```python
from sklearn import linear_model
```

```python
WalmartLinear = linear_model.LinearRegression(normalize=True)
WalmartLinear.fit(XTrain, YTrain)
```

```python
WalmartLinear.coef_
```

```python
Test= WalmartLinear.predict(XTest)
```

```python
plt.plot(YTest, Test,'go')
plt.plot(YTest, YTest,'r-')
plt.show()
```

```python
wmLinear = linear_model.LinearRegression(normalize=True)
wmLinear.fit(XTrain, YTrain)
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
```

```python
alphas = np.linspace(10, 20, 10)
alphas
```

```python
testError = np.empty(10)

for i, alpha in enumerate(alphas) :
    
    lasso = Lasso(alpha=alpha)
    lasso.fit(XTrain, YTrain)
    testError[i] = mean_squared_error(YTest, lasso.predict(XTest))
```

```python
plt.plot(alphas, testError, 'r-')
plt.show()
```
